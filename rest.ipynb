```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import jensenshannon as jsd
```

```python
df = pd.read_csv('data.csv')
```

```python
v = sorted(list(set(df[['grid_id']].to_numpy().flatten())))
n = len(v)

print(v)
print(n)
```

```python
id_rest = [] 

for i, grid_id in enumerate(v):
    print(f'i = {i}')
    tmp1 = len(df[(df['道路型態大類別名稱'] == '交岔路') & (df['號誌-號誌種類名稱'] == '無號誌') & (df['grid_id'] == grid_id)])
    tmp2 = len(df[(df['道路型態大類別名稱'] == '交岔路') & (df['號誌-號誌種類名稱'] != '無號誌') & (df['grid_id'] == grid_id)])
    tmp3 = len(df[(df['道路型態大類別名稱'] != '交岔路') & (df['號誌-號誌種類名稱'] == '無號誌') & (df['grid_id'] == grid_id)])
    tmp4 = len(df[(df['道路型態大類別名稱'] != '交岔路') & (df['號誌-號誌種類名稱'] != '無號誌') & (df['grid_id'] == grid_id)])
    id_rest.append([tmp1, tmp2, tmp3, tmp4])
```

```python
id_rest = np.array(id_rest)
id_rest
```

```python
np.save('id_rest.npy', id_rest)
```

```python
id_rest = np.load('id_rest.npy')
id_rest
```

```python
adj_matrix = np.zeros((n, n))

for i in range(n):
    print(f'i = {i}')
    for j in range(i, n): # adj_matrix is symmetric (adj_matrix[i][j] == adj_matrix[j][i])
        ri = id_rest[i] / sum(id_rest[i])
        rj = id_rest[j] / sum(id_rest[j])
        # Jensen-Shannon distance is the square root of the Jensen-Shannon divergence
        adj_matrix[i][j] = 1 - jsd(ri, rj) ** 2 
```

```python
tmp = adj_matrix.copy()

for i in range(n):
    for j in range(n):
        if i > j: tmp[i][j] = tmp[j][i]
            
tmp
```

```python
np.save('adj_matrix_rest.npy', tmp)
```
