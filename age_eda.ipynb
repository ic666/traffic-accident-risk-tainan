```python
import pandas as pd
```

```python
all_csv_with_grid = pd.read_csv("all_csv_with_grid.csv",engine="python")
```

```python
all_csv_with_grid.sort_values(["months", "time slot"])
```

```python
all_csv_with_grid_age = all_csv_with_grid.drop(all_csv_with_grid[all_csv_with_grid["當事者事故發生時年齡"]==-1].index)
all_csv_with_grid_age
```

```python
bins= [0,12,65,120]
labels = ['kid','adult','elder']
all_csv_with_grid_age['age_group'] = pd.cut(all_csv_with_grid_age['當事者事故發生時年齡'], bins=bins, labels=labels, right=False)
all_csv_with_grid_age
```

```python
all_csv_with_grid_age[:10]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
all_csv_with_grid_age[all_csv_with_grid_age["age_group"]=='elder'].cause.value_counts().plot.barh(title='age_group:elder')
```

```python
all_csv_with_grid_age[all_csv_with_grid_age["age_group"]=='adult'].cause.value_counts().plot.barh(title='age_group:adult')
```
