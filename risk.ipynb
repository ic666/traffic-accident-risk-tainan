```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import jensenshannon as jsd
```

```python
df = pd.read_csv('data.csv')
```

```python
v = np.array(sorted(list(set(df[['grid_id']].to_numpy().flatten()))))
n = len(v)

print(v)
print(n)
```

```python
id_accidents = [[len(df[(df['grid_id'] == node) & (df['time slot'] == i)]) for i in range(24)] for node in v]
id_accidents = np.array(id_accidents)
id_accidents # id_accidents[i][j]: number of accidents of 'id v[i]' at 'time slot j' 
```

```python
np.save('id_accidents.npy', id_accidents)
```

```python
id_accidents = np.load('id_accidents.npy')
id_accidents
```

```python
adj_matrix = np.zeros((n, n))

for i in range(n):
    print(f'i = {i}')
    for j in range(i, n): # adj_matrix is symmetric (adj_matrix[i][j] == adj_matrix[j][i])
        ri = id_accidents[i] / sum(id_accidents[i])
        rj = id_accidents[j] / sum(id_accidents[j])
        # Jensen-Shannon distance is the square root of the Jensen-Shannon divergence
        adj_matrix[i][j] = 1 - jsd(ri, rj) ** 2 
```

```python
np.save('adj_matrix_risk.npy', adj_matrix)
```

```python
tmp = adj_matrix.copy()

for i in range(n):
    for j in range(n):
        if i > j: tmp[i][j] = tmp[j][i]
            
tmp
```

```python
np.save('adj_matrix_risk.npy', tmp)
```
